```python
import pandas as pd
```

```python
#Importing All Files
data=pd.read_csv('C:\\Users\\Ahnaf\\Desktop\\LHL Final Project\\LHL-Final-Project-Global-Airport-Analysis-\\data.csv')
world_data=pd.read_csv('C:\\Users\\Ahnaf\\Desktop\\LHL Final Project\\LHL-Final-Project-Global-Airport-Analysis-\\world-data-2023 cleaned.csv')
airport_data=pd.read_csv('C:\\Users\\Ahnaf\\Desktop\\LHL Final Project\\LHL-Final-Project-Global-Airport-Analysis-\\airports_data uncleaned.csv')
```

```python
# 1. Airport Distribution Analysis
#    - Evaluate the distribution of airports across countries.
```

```python
#Evaluate the distribution of airports across countries

airport_country_distribution=airport_data['country'].value_counts()
airport_country_distribution.head(10)
```

```python
#Data Visualization for airport_country_distribution

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top_country=airport_country_distribution.head(10)
plt.figure(figsize=(12,7))
sns.barplot(x=top_country.index,y=top_country.values)
plt.title('Top 10 Countries by Number of Airports')
plt.xlabel('Country')
plt.ylabel('Number of Airports')
```

```python
airport_data.head()
```

```python
############################Distribution of airport types (small, medium, large) ###########################################

## Define altitude thresholds for categorization

lower_threshold=airport_data['alt'].quantile(0.33)
upper_threshold=airport_data['alt'].quantile(0.66)

# Categorize airports

airport_data['type']='Small'
airport_data.loc[(airport_data['alt']> lower_threshold) & (airport_data['alt']<=upper_threshold), 'type']='Medium'
airport_data.loc[airport_data['alt']>upper_threshold, 'type']='Large'

# Count the number of each type of airport for each country

airport_type_distribution=airport_data.groupby('country')['type'].value_counts().unstack().fillna(0)

# Get the top countries by total number of airports

top_countries_by_airports=airport_type_distribution.sum(axis=1).nlargest(10).index

# Subset the data for these top countries

top_countries_distribution = airport_type_distribution.loc[top_countries_by_airports]


# Plotting the distribution of airport types for the top countries
plt.figure(figsize=(14, 8))
top_countries_distribution.plot(kind='bar', stacked=True, colormap='viridis', figsize=(14,8))
plt.title('Distribution of Airport Types in Top 10 Countries by Number of Airports')
plt.xlabel('Country')
plt.ylabel('Number of Airports')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2.Economic Analysis
#    - Correlate GDP or GDP per capita with the number of airports.
#    - Explore countries with high GDP but fewer airports.
```

```python
# Grouping the airport data by country to get the total number of airports for each country

airport_counts=airport_data.groupby('country').size().reset_index(name='number_of_airport')
#airport_counts.head()

#Merging tow csv files Airport.CSV

# Adjusting the merge to use the correct column names
merged_data = pd.merge(airport_counts, world_data[['Country', 'GDP']], left_on='country', right_on='Country', how='inner')
# Dropping duplicate 'Country' column
merged_data.drop('Country', axis=1, inplace=True)  
merged_data.head()
```

```python
#Correlation Analysis:

# Converting GDP column to numeric after removing commas and dollar signs
merged_data['GDP'] = merged_data['GDP'].str.replace(',', '').str.replace('$', '').astype(float)
merged_data['GDP']
```

```python
#Correlation Analysis:

correlation_gdp_airport=merged_data['GDP'].corr(merged_data['number_of_airport'])
correlation_gdp_airport

# 0.82, which indicates a strong positive linear relationship. 
        #This suggests that, generally,countries with higher GDPs tend to have more airports.
```
