```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing All Files
data=pd.read_csv('C:\\Users\\Ahnaf\\Desktop\\LHL Final Project\\LHL-Final-Project-Global-Airport-Analysis-\\data.csv')
world_data=pd.read_csv('C:\\Users\\Ahnaf\\Desktop\\LHL Final Project\\LHL-Final-Project-Global-Airport-Analysis-\\world-data-2023.csv')
airport_data=pd.read_csv('C:\\Users\\Ahnaf\\Desktop\\LHL Final Project\\LHL-Final-Project-Global-Airport-Analysis-\\airports_data.csv')
```

```python
# Airport Altitude Analysis
#    - Analyze the distribution of airport altitudes.
#    - Identify countries with airports at different altitudes and analyze reasons.
```

```python
# Analyze the distribution of airport altitudes.

plt.figure(figsize=(12,7))
sns.histplot(airport_data['alt'], bins=50,kde=True)
```

```python
# Grouping by country and calculating the average, minimum, and maximum altitudes
altitudes_by_country=airport_data.groupby('country').agg({'alt':['mean','min','max']}).reset_index()

# Sorting countries by average altitude to find top countries with highest average altitudes

top_countries_by_altitude=altitudes_by_country.sort_values(by=('alt','mean'), ascending=False).head(10)
top_countries_by_altitude.columns = ['Country', 'Mean Altitude', 'Min Altitude', 'Max Altitude']
top_countries_by_altitude
```

```python
# Plotting the data for better visualization

plt.figure(figsize=(15,10))

# Bar plot for average altitude
sns.barplot(x='Mean Altitude',y='Country',data=top_countries_by_altitude)

# Aesthetics
plt.title('Top 10 Countries by Average Airport Altitude', fontsize=16)
plt.xlabel('Altitude (in feet)', fontsize=14)
plt.ylabel('Country', fontsize=14)
#plt.legend()
plt.grid(axis='x')
plt.show()
```
