```python
#Merged Data, Name : Data.CSV

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
columns=['lat', 'lon', 'alt', 'Latitude', 'Longitude']
for i, col in enumerate (columns,1):
    plt.subplot(2,3,i)
    sns.histplot(data[col],bins=30,kde=True)
    plt.title(f'Distributions of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Distribution of Airport Altitude: (Observations)

# Most airports are located at relatively lower altitudes.
# There's a significant peak near the sea level.
# There are also some airports located at very high altitudes, as indicated by the long tail of the distribution.
# Distribution of Airports by Latitude:

# The distribution of airports by latitude shows that there are concentrations of airports near the equator and in the northern hemisphere.
# The southern hemisphere has fewer airports compared to the northern hemisphere, especially as you move further south.
# Distribution of Airports by Longitude:

# The distribution is relatively uniform with slight peaks around certain regions. This might be due to concentrations of countries or populated areas.
```

```python
# Correlation Analysis, Data CSV 

# Computing the correlation matrix
correlation_matrix = data[['lat', 'lon', 'alt']].corr()

# Plotting the heatmap for the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix for Numeric Columns')
plt.show()
```
